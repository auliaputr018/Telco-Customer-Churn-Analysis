# 1. Import Library

```python
import pandas as pd
import pickle
```

# 2. Load Data

```python
# load model terbaik
with open('best_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

# 3. Prediction 1

```python
# membuat data inference baru
data_inf = pd.DataFrame({

    'gender': ['Male'],
    'SeniorCitizen': [0],
    'Partner': ['Yes'],
    'Dependents': ['No'],
    'tenure': [12],
    'PhoneService': ['Yes'],
    'MultipleLines': ['No'],
    'InternetService': ['Fiber optic'],
    'OnlineSecurity': ['No'],
    'OnlineBackup': ['Yes'],
    'DeviceProtection': ['No'],
    'TechSupport': ['No'],
    'StreamingTV': ['Yes'],
    'StreamingMovies': ['Yes'],
    'Contract': ['Month-to-month'],
    'PaperlessBilling': ['Yes'],
    'PaymentMethod': ['Electronic check'],
    'MonthlyCharges': [85.5],
    'TotalCharges': [1026.0]

})

# tampilkan data inference
data_inf
```

```python
# prediksi churn
pred_inf = loaded_model.predict(data_inf)

# probabilitas churn
pred_prob = loaded_model.predict_proba(data_inf)

print("Prediction :", pred_inf[0])
print(f"Probability Non-Churn : {pred_prob[0][0]:.2%}")
print(f"Probability Churn     : {pred_prob[0][1]:.2%}")
```

```python
# interpretasi hasil prediksi
if pred_inf[0] == 1:
    print("Customer diprediksi akan churn")
else:
    print("Customer diprediksi tidak churn")
```

Insight:

>- Berdasarkan hasil inference, model memprediksi bahwa pelanggan akan melakukan churn dengan probabilitas sebesar 78.16%. Hal ini menunjukkan bahwa model memiliki tingkat keyakinan yang cukup tinggi terhadap prediksi churn yang dihasilkan.
>
>- Prediksi tersebut dipengaruhi oleh beberapa karakteristik pelanggan yang sebelumnya juga terlihat memiliki hubungan dengan churn pada tahap EDA dan feature importance. Pelanggan menggunakan kontrak bulanan (month-to-month), metode pembayaran electronic check, serta tidak menggunakan layanan tambahan seperti OnlineSecurity dan TechSupport.
>
>- Selain itu, pelanggan juga memiliki tenure yang masih relatif rendah, yaitu 12 bulan, yang menunjukkan bahwa pelanggan belum memiliki loyalitas yang kuat terhadap layanan perusahaan. Kombinasi faktor-faktor tersebut menyebabkan pelanggan memiliki risiko churn yang cukup tinggi menurut model yang telah dibuat.
>
>- Pelanggan dengan karakteristik seperti ini dapat menjadi prioritas utama dalam strategi retensi karena memiliki risiko churn yang cukup tinggi menurut model.
